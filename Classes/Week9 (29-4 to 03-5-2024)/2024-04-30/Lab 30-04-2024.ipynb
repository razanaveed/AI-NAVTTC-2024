```python
import re
import string

def clean_text(text):
    text=text.lower()
    text = re.sub(r'\d+','',text)
    text = text.translate(text.maketrans('','',string.punctuation))
    return text
```

```python
text_original="A+ Quick45, Brow/n Fox Jum=ps over! a little and& Lazy? Dog2"
```

```python
#clean
text=clean_text(text_original)
print(text)
```

```python
!pip install nltk
```

```python
#tokenization
from nltk import word_tokenize
import nltk
nltk.download('punkt')
```

```python
words=word_tokenize(text)
print(words)
```

```python
from nltk import sent_tokenize
```

```python
sentences=sent_tokenize(text_original)
print(sentences)
```

```python
#POS Tag
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
pos_tags= pos_tag(words)
print(pos_tags)
```

```python
from collections import Counter
```

```python
pos_cat = [tag[1] for tag in pos_tags]
print(pos_cat)
```

```python
pos_count=Counter(pos_cat)
print(pos_count)
```

```python
!pip install matplotlib
```

```python
#Plot POS Distribution
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.bar(pos_count.keys(),pos_count.values())
plt.title('POS Distribution')
plt.xlabel('POS Tags')
plt.ylabel('Count')
plt.show()
```
