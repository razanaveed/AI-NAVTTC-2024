```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("tv_shows.csv")
```

```python
df.dropna(inplace=True)
```

```python
# Create a DataFrame from the value counts
year_counts = df["Year"].value_counts().reset_index()
#year_counts.columns = ["Year", "Count"]
year_counts
```

```python
# Reshape the data
X = year_counts["Year"].values.reshape(-1, 1)  # Independent variable (Year)
X
```

```python
# Reshape the data
Y = year_counts["count"].values  # Dependent variable (Count)
Y
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
year_2024 = np.array([[2450]])
predicted_count=model.predict(year_2024)
predicted_count
```
