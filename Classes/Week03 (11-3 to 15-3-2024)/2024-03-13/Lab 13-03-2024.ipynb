```python
import random
class Vehicle:
    name="Vehicle"
    def start_engine(self):        
        print(f"Start your {self.name} engine!")
    def accelerate(self):
        print(f"You're moving {self.name} slowly, speed up!")
    def brake(self):
        print(f"You're moving {self.name} very fast, apply brake immediately!")
    def Work(self,speed):
        if (speed==0):
            return self.start_engine()
        elif (speed<50):
            return self.accelerate()
        elif (speed>80):
            return self.brake()
        else:
            return print(f"You're moving {self.name} at {speed}/mph")

        
class FourWheel(Vehicle):
    name="Four Wheel"

class Car(FourWheel):
    name="Car"
    
car1= Car()
#bike1=Bike()
car1.Work(0)
print("")
temp=0
for i in range(1,11):
    speed=random.randint(0, 100)
    print(f"{'Increasing' if speed>=temp else 'Decreasing'} Speed: {speed}")
    #car1.Work(speed)
    car1.Work(speed)
    print("")
    temp=speed
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # Abstract method, to be overridden by subclasses


class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"


class Labrador(Dog):
    def speak(self):
        return f"{self.name} says Bark!"


# Creating instances
a = Animal("Generic Animal")
d = Dog("Buddy")
l = Labrador("Max")

# Calling methods
print(a.speak())  # Output: None (from Animal class)
print(d.speak())  # Output: Buddy says Woof!
print(l.speak())  # Output: Max says Bark!
```

```python
class SMS:
    def do_sms(self):
        return "I can send SMS"

class Call:
    def do_call(self):
        return "I can make call"

class Email:
    def do_email(self):
        return "I can send email"

class Camera:
    def do_click(self):
        return "I can take selfie"

class Browsing:
    def do_surf(self):
        return "I can surf internet"

class Game:
    def do_play(self):
        return "I can run games"

class Mobile:
    def __init__(self,name):
        self.name=name
    def __str__(self):
        return f"I am {self.name}"

class Symbian(Mobile,SMS,Call):
    pass

class Blackberry(Mobile,SMS,Call,Camera,Email,Browsing):
    pass

class Android(Mobile,SMS,Call,Camera,Email,Browsing,Game):
    pass

nokia=Symbian("Nokia-3310")
bb=Blackberry("KeyOne")
samsung=Android("Galaxy A-32")

print(f"{nokia}, {nokia.do_sms()},{nokia.do_call()}")
print(f"{bb}, {bb.do_sms()},{bb.do_call()},{bb.do_email()},{bb.do_click()},{bb.do_surf()}")
print(f"{samsung}, {samsung.do_sms()},{samsung.do_call()},{samsung.do_play()}")
```
