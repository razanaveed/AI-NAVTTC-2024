```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#Generate Syntethic Data
x, y = make_blobs(n_samples=500, centers=6, cluster_std=0.6, random_state=0)
```

```python
x
```

```python
y
```

```python
kmeans = KMeans(n_clusters=6)
```

```python
labels = kmeans.fit(x)
labels = kmeans.predict(x)
```

```python
labels
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100, c='red',marker='X', label='centroids')
plt.title("Kmeans cluster")
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris= load_iris()
x=iris.data
y=iris.target
```

```python
pca=PCA(n_components=3)
x_pca=pca.fit_transform(x)
```

```python
plt.scatter(x_pca[:,0],x_pca[:,2],c=y,cmap='viridis')
plt.title("PCA of IRIS")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```
