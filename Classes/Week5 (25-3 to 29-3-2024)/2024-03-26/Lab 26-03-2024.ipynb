```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('marks.csv')
#df=df.sort_values(by="Marks")
print(df)
#Line Chart Age/Marks
plt.figure(figsize=(8, 5))
plt.plot(df['Age'], df['Marks'], marker='o', linestyle='-')
plt.title('Marks By Age')
plt.xlabel('Age')
plt.ylabel('Marks')
plt.grid(True)
plt.show()

#Histogram for Grade
plt.figure(figsize=(8, 5))
plt.hist(df['Grade'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='Age', color='lightblue')
plt.title('Age Violin Plot')
plt.xlabel('Age')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='Name',y='Age', color='lightblue')
plt.title('Name Violin Plot')
plt.xlabel('Name')
plt.ylabel('Age')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('dataset.csv')
#df=df.sort_values(by="Marks")
print(df)

#Line Chart chol By Age
plt.figure(figsize=(8, 5))
plt.plot(df['age'], df['chol'], marker='o', linestyle='-')
plt.title('Line Chart: chol By Age')
plt.xlabel('age')
plt.ylabel('chol')
plt.grid(True)
plt.show()

# Bar Chart for trestbps by age
plt.figure(figsize=(8, 5))
plt.bar(df['age'], df['trestbps'], color='skyblue')
plt.title('Bar Chart: trestbps by age')
plt.xlabel('age')
plt.ylabel('trestbps')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Scatter plot thalach by cp
plt.figure(figsize=(8, 5))
plt.scatter(df['cp'], df['thalach'], color='orange', marker='o')
plt.title('Scatter Plot: thalach by cp')
plt.xlabel('cp')
plt.ylabel('thalach')
plt.grid(True)
plt.show()

# Scatter plot thalach by cp
plt.figure(figsize=(8, 5))
plt.scatter( df['chol'],df['thalach'], color='orange', marker='o')
plt.title('Scatter Plot: thalach  by chol')
plt.xlabel('chol')
plt.ylabel('thalach')
plt.grid(True)
plt.show()
```
