#**Line Plot**
Used to visualize the relationship between two continuous variables over a continuous interval.

#**Scatter Plot**
Shows the relationship between two variables by displaying points in a two-dimensional space.

#**Bar Plot**
Displays categorical data with rectangular bars, where the length of each bar represents the value of the category.

#**Histogram**
Represents the distribution of a continuous variable by dividing the data into bins and displaying the frequency of observations in each bin.

#**Box Plot (Box-and-Whisker Plot)**
Provides a graphical summary of the distribution of a continuous variable, including measures of central tendency and variability.

#**Violin Plot**
Combines aspects of a box plot and a kernel density plot to provide a more comprehensive summary of the distribution of a continuous variable.

#**Heatmap**
Visualizes the magnitude of a variable as colors in a matrix, where each cell represents the intersection of two categorical variables.

#**Pie Chart**
Represents the proportions of different categories in a dataset as sectors of a circle.

#**Stacked Bar Plot**
Displays multiple sets of data as stacked bars within each category.

#**Area Plot**
Shows the evolution of a variable over time by filling the area below the line connecting data points.

#**Pair Plot (Pairwise Scatter Plot)**
Displays pairwise relationships between different variables in a dataset.

#**Correlation Heatmap**
Visualizes the pairwise correlation coefficients between variables in a dataset as colors in a matrix.

#**Joint Plot**
Combines scatter plots with histograms or kernel density plots to visualize the relationship between two variables and their individual distributions.

#**Facet Grid Plot**
Divides the data into subsets based on the values of one or more categorical variables and creates separate plots for each subset.

#**Radar Chart (Spider Plot)**
Displays multivariate data on a two-dimensional plane with each variable represented as a spoke from the center.

#**3D Plot**
Visualizes three-dimensional data in a three-dimensional space using surfaces, scatter points, or other geometric shapes.


#**Stream Plot**
Visualizes the flow of a vector field using streamlines.

#**Contour Plot**
Represents the topographical surface of a three-dimensional function as contours in a two-dimensional space.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Read the heart.csv file into a DataFrame
heart_data = pd.read_csv("heart.csv")

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(heart_data['age'], heart_data['chol'])
plt.title('Line Plot: Age vs. Cholesterol', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Cholesterol', fontsize=14)
plt.grid(True)
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(heart_data['age'], heart_data['thalach'])
plt.title('Scatter Plot: Age vs. Maximum Heart Rate', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Maximum Heart Rate', fontsize=14)
plt.grid(True)
plt.show()

# 3. Bar Plot
plt.figure(figsize=(10, 6))
heart_data['sex'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Bar Plot: Gender Distribution', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)
plt.show()

# 4. Histogram
plt.figure(figsize=(10, 6))
plt.hist(heart_data['age'], bins=10, color='skyblue', alpha=0.7)
plt.title('Histogram: Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.show()

# 5. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', data=heart_data)
plt.title('Box Plot: Age by Gender', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.grid(True)
plt.show()

# 6. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='age', data=heart_data)
plt.title('Violin Plot: Age by Gender', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.grid(True)
plt.show()

# 7. Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heart_data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap: Correlation Matrix', fontsize=16)
plt.show()

# 8. Pie Chart
plt.figure(figsize=(8, 6))
heart_data['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Pie Chart: Gender Distribution', fontsize=16)
plt.ylabel('')
plt.show()

# 9. Stacked Bar Plot
plt.figure(figsize=(10, 6))
pd.crosstab(heart_data['sex'], heart_data['target']).plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])
plt.title('Stacked Bar Plot: Heart Disease by Gender', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)
plt.show()

# 10. Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(heart_data.index, heart_data['age'], color='skyblue', alpha=0.4)
plt.title('Area Plot: Age Distribution', fontsize=16)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.grid(True)
plt.show()

# 11. Pair Plot
# Pairplot of selected features
plt.figure(figsize=(10, 8))
selected_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
sns.pairplot(heart_data[selected_features], diag_kind='kde', palette='viridis')
plt.suptitle('Pairplot of Selected Features', fontsize=16)
plt.show()

# 12. Correlation Heatmap (repeated for different libraries)
plt.figure(figsize=(10, 8))
correlation_matrix = heart_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Correlation Matrix', fontsize=16)
plt.show()

# 13. Joint Plot
sns.jointplot(x='age', y='thalach', data=heart_data, kind='scatter')
plt.suptitle('Joint Plot: Age vs. Maximum Heart Rate', fontsize=16)
plt.show()

# 14. Facet Grid Plot
g = sns.FacetGrid(heart_data, col='sex', hue='target', height=5)
g.map(plt.scatter, 'age', 'chol')
g.add_legend()
plt.suptitle('Facet Grid Plot: Age vs. Cholesterol by Gender', fontsize=16)
plt.show()

# 15. Radar Chart
from math import pi

labels=np.array(['age', 'sex', 'cp', 'trestbps', 'chol'])
stats=heart_data[labels].mean().values

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()

stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='skyblue', alpha=0.25)
ax.plot(angles, stats, color='skyblue', linewidth=2)
ax.set_yticklabels([])
plt.title('Radar Chart: Average Values of Selected Features', fontsize=16)
plt.show()

# 17. 3D Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(heart_data['age'], heart_data['thalach'], heart_data['chol'], c=heart_data['target'], cmap='coolwarm', s=50)
ax.set_xlabel('Age')
ax.set_ylabel('Maximum Heart Rate')
ax.set_zlabel('Cholesterol')
plt.title('3D Plot: Age, Maximum Heart Rate, and Cholesterol', fontsize=16)
plt.show()

# 18. Tree Plot
# Assuming you have a decision tree model trained using heart_data

# 19. Stream Plot
# Stream Plot
x = np.linspace(0, 10, 10)  # Adjusting the number of points for simplicity
y = np.linspace(0, 10, 10)
X, Y = np.meshgrid(x, y)
u = np.sin(X)
v = np.cos(Y)

plt.figure(figsize=(10, 6))
plt.streamplot(X, Y, u, v, color='skyblue')
plt.title('Stream Plot: Vector Field', fontsize=16)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# 20. Contour Plot
x = np.linspace(0, 10, 10)  # Adjusting the number of points for simplicity
y = np.linspace(0, 10, 10)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, cmap='coolwarm')
plt.title('Contour Plot: 3D Function', fontsize=16)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.colorbar()
plt.show()
```
