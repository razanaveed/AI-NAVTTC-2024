```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPool2D
from keras.datasets import mnist
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train,y_train),(X_test,Y_test) = mnist.load_data()
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
plt.imshow(X_train[5])
```

```python
# Define the number of rows and columns for the subplot
rows = 2
cols = 5

# Create a new figure
plt.figure(figsize=(10, 5))

# Plot each image in a subplot
for i in range(rows * cols):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(X_train[i])
    plt.axis('off')  # Turn off axis
    plt.title(f"Image {i+1}")  # Add title

plt.show()
```

```python
#reshape
X_Train= X_train.reshape(60_000,28,28,1) # 1 = channel = Gray scale
X_test= X_test.reshape(10_000,28,28,1) # 1 = channel = Gray scale
```

```python

#one_hot_encode

y_train_one_hot=to_categorical(y_train)
y_test_one_hot=to_categorical(Y_test)
```

```python
print(y_train_one_hot[11],y_train[11])
```

```python
plt.imshow(X_train[5])
```

```python
#Build CNN Model
model=Sequential()
#addd model layers
model.add(Conv2D(128,kernel_size=3,activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(64,kernel_size=3,activation='relu'))
model.add(Conv2D(32,kernel_size=3,activation='relu'))
model.add(MaxPool2D(pool_size=(2,2),strides=None,padding='valid',data_format=None))
model.add(Flatten())
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train_one_hot,validation_data=(X_test,y_test_one_hot),epochs=10)
```
